### 1. Bestimmung der Haftreibung

Zur Messung der Haftreibung stellten wir das Messgerät auf das Buch, und erhöhten langsam und gleichmäßig den
Winkel zwischen dem Buch und der horizontalen Ebene, bis es anfing zu rutschen. Das Experiment wurde 6 Mal zur Bestimmung des Wertes und des wahrscheinlichen Fehlers der Haftreibung $\mu_H$ wiederholt. Wir kennen aus der Theorie, dass für den Haftreibungskoeffizienten gilt

\begin{equation}
    \mu_H=\tan\alpha_H,
\end{equation}

wobei $\alpha_H$ der Winkel beim Losrutschen darstellt. 

Eine erste Abschätzung dieses Grenzwinkels, $\alpha_{H_0}$, konnte bereits im Vorexperiment mit dem Neigungsmesser gemessen werden: nach 10 Messungen stellte sich heraus, dass $\overline{\alpha_{H_0}}+\Delta\overline{\alpha_{H_0}} = (15,6 \pm 0,1)^{\textrm{o}}$. Der tatsächliche Wert von $\alpha_H$ ließ sich aber schnell durch eine Auswertung der Komponenten des Beschleunigungsvektors $(a_x, a_y, a_z)$ bezüglich der Senkrechten ermitteln (siehe **Fig. 2**).

![](https://raw.githubusercontent.com/allgebrist/studienkolleg-physik-praktikum/master/Bilder/8.2.jpg)

Der Winkel $\alpha_H$ ist gleich dem Winkel $\psi$ zwischen der $Y$-Achse des Messgeräts und dem Gravitationssystem [1]. Der Winkel der schiefen Ebene kann man unter Verwendung der Beschleunigungsmessungen $a_x$, $a_y$ und $a_z$ im Ruhezustand bzw. beim Losrutschen wie folgt berechnen:

\begin{equation}
    \alpha_H = \psi = arctan\left(\frac{a_y}{\sqrt{a_x^2+a_z^2}}\right).
\end{equation}

Mithilfe dieser Gleichung erstellten wir eine neue Spalte in unseren Tabellen, um die Änderung des Winkels im Laufe des Experiments nachvollziehen zu können. Ein Blick auf unsere Messungen finden Sie weiter unten im Dokument.

Der wahrscheinliche Fehler von $\mu_H$ kann nach Gauß folgendermaßen bestimmt werden:

\begin{aligned}
    \Delta\mu_H &= \sqrt{\left[\frac{\partial}{\partial a_x}\left(\frac{a_y}{\sqrt{a_x^2+a_z^2}}\right)\cdot\Delta a_x\right]^2 
                            + \left[\frac{\partial}{\partial a_y}\left(\frac{a_y}{\sqrt{a_x^2+a_z^2}}\right)\cdot\Delta a_y\right]^2  
                            + \left[\frac{\partial}{\partial a_z}\left(\frac{a_y}{\sqrt{a_x^2+a_z^2}}\right)\cdot\Delta a_z\right]^2 } \\ 
                   &= \sqrt{\left(-\frac{a_x a_y}{(a_x^2+a_z^2)^{3/2}}\cdot\Delta a_x\right)^2 
                            + \left(\frac{1}{\sqrt{a_x^2+a_z^2}}\cdot\Delta a_y\right)^2
                            + \left(-\frac{a_y a_z}{(a_x^2+a_z^2)^{3/2}}\cdot\Delta a_z\right)^2 }.
\end{aligned}

```python
ax_mittel = np.mean(tabelle1["Acceleration x (m/s^2)"][600:1000])
ay_mittel = np.mean(tabelle1["Acceleration y (m/s^2)"][600:1000])
az_mittel = np.mean(tabelle1["Acceleration z (m/s^2)"][600:1000])

ax_std = np.std(tabelle1["Acceleration x (m/s^2)"][600:1000], ddof = 1)
ay_std = np.std(tabelle1["Acceleration y (m/s^2)"][600:1000], ddof = 1)
az_std = np.std(tabelle1["Acceleration z (m/s^2)"][600:1000], ddof = 1)

ax_err = ax_std / math.sqrt(400)
ay_err = ay_std / math.sqrt(400)
az_err = az_std / math.sqrt(400)

wahr_fehler = math.sqrt( (((ax_mittel*ay_mittel)/(ax_mittel**2 + az_mittel**2)**(3/2))*ax_err)**2 
                        + ((1/(ax_mittel**2 + az_mittel**2)**(1/2))*ay_err)**2 
                        + (((ay_mittel*az_mittel)/(ax_mittel**2 + az_mittel**2)**(3/2))*az_err)**2 )
```

### 2. Bestimmung der Gleitreibung

Starten Sie die Messung mit ca. 5 verschiedenen Winkeln, die größer sind als der Grenzwinkel. „Beobachten“ Sie die Bewegung und werten Sie ca. 5 „gute Messungen“ aus. Durch die Beschleunigungsmessung können Sie sehr präzise die Zeitdauer der Beschleunigung messen. Die Bewegung sollte gleichmäßig beschleunigt ablaufen und die Werte (und Fehler) von der Gleitreibung µG nicht vom Winkel abhängen.

die Länge der Ebene $x=(26,5\pm 0,1)\textrm{ cm}$

\begin{equation}
    \mu_G=\frac{a_y}{g \cos\psi}.
\end{equation}

Da der Wert von $\psi = \alpha_H$ schon im letzten Abschnitt ermittelt wurde, sind wir nun in der Lage, ihn in diese Gleichung einzusetzen. 

\begin{aligned}
    t\pm \delta t &= \sqrt{\frac{2x}{a}} \pm \frac{1}{2}\left(\delta x + \delta a\right) \\
                  &= \sqrt{\frac{2x}{a}} \pm \frac{1}{2}\left(\frac{\Delta x}{x} + \frac{\Delta a}{a}\right)
\end{aligned}
